```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import ipywidgets as widgets
```

### Importing data into dataframe

```
df = pd.read_csv('Video_Games_Sales.csv')
df
```

```
df.describe()
```

### Dropping some data

```
df = df.drop(df[df['Global_Sales'] == 82.74].index) # drop WiiSports, game sold with WII
df = df.drop(df[df['Name'] == 'Strongest Tokyo University Shogi DS'].index) # drop anomaly, DS game in 1985
df = df[df['Name'].notna()]
df = df[df['Genre'].notna()]
df = df[df['Platform'].notna()]
df = df[df['Publisher'].notna()]
df = df[df['Year_of_Release'].notna()]
df = df.drop(df[df['Year_of_Release'] == 2017].index) 
df = df.drop(df[df['Year_of_Release'] == 2020].index) 
```

```
df.describe()
```

```
fig1, ax = plt.subplots(figsize=(30,5))
sns.violinplot(x="Global_Sales", y="Genre", data = df, ax=ax)
```

```
#TODO: Remplacer 1993 par une variable qu'on adaptera en selection sur Dashboard
style = {'description_width': 'initial'}

limit_case = widgets.IntSlider(
    value=1993,
    min=1980,
    max=2020,
    step=1,
    description='Year of Release:',
    disabled=False,
    style=style)

def update_df(year_i):
    df_i = df[df['Year_of_Release'] == year_i]

widgets.interactive(update_df, year_i=limit_case)

df_i = df[df['Year_of_Release'] == 1993]

fig2, ax = plt.subplots(figsize=(20,30))
fig2 = sns.barplot(y = df_i['Name'], x = df_i['Global_Sales'], ax=ax)
plt.xlabel("Name")
plt.ylabel("Global_Sales")
```

```
fig2 = sns.barplot(x = df['Genre'].value_counts(), y = df['Genre'].value_counts().keys(), palette="flare")
plt.xlabel("Number of games")
plt.ylabel("Genre")
```

```
table_platforms = pd.pivot_table(df,values=['Global_Sales'],index=['Year_of_Release'],columns=['Platform'],aggfunc='count',margins=False)

plt.figure(figsize=(19,16))
sns.heatmap(table_platforms['Global_Sales'],linewidths=.5,annot=True,fmt='2.0f',vmin=0)
plt.title('Count of games')
```

```
corrmat = df.corr()
sns.heatmap(corrmat, vmax=.8, square=True);
```

```
x = df['Year_of_Release'].value_counts().keys()
y = df['Year_of_Release'].value_counts()
plt.figure(figsize=(15,3))
plt.bar(x,y)
plt.ylabel("Number of games")
plt.xlabel("Year of Release")


plt.show()

# Je pense couper le DF en 2 parties, les annees 80-93 et 93-2020
```

```
data = [df['Year_of_Release'].groupby(df['Year_of_Release']).mean(),
df['EU_Sales'].groupby(df['Year_of_Release']).sum(), 
df['NA_Sales'].groupby(df['Year_of_Release']).sum(),
df['JP_Sales'].groupby(df['Year_of_Release']).sum(),
df['Other_Sales'].groupby(df['Year_of_Release']).sum()
]

headers = ["Year_of_Release","EU_Sales", "NA_Sales", "JP_Sales", "Other_Sales"]

dfpie = pd.concat(data, axis=1, keys=headers)
dfpie
```

```
fig, axes = plt.subplots(6, 6, figsize=(20, 36))

for i, (idx, row) in enumerate(dfpie.set_index('Year_of_Release').iterrows()):
    ax = axes[i // 6, i % 6]
    row = row[row.gt(row.sum() * .01)]
    ax.pie(row, labels=row.index, startangle=30, autopct='%1.1f%%')
    ax.set_title(idx)

fig.subplots_adjust(wspace=0.2)
```
